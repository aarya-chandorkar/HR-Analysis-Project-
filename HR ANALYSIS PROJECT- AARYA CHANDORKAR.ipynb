# Visualization of Employee Distribution

```python
df.JobRole.unique()
```

```python
df.Department.unique()
```

```python
#groupby('Department') → Groups rows by Department
#['JobRole'].count() → Counts how many employees have a JobRole in each department
df.groupby('Department')['JobRole'].count()
```

```python
# Group by Department and JobRole to count employees
dept_counts = df['Department'].value_counts()
jobrole_counts = df['JobRole'].value_counts()

# Create subplots
plt.figure(figsize=(12, 6))

# First subplot for Department distribution
plt.subplot(1, 2, 1)  # (rows, columns, position)
sb.barplot(x=dept_counts.index, y=dept_counts.values)
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Distribution Across Departments')
plt.xticks(rotation=30)

# Second subplot for Job Role distribution
plt.subplot(1, 2, 2)
sb.barplot(x=jobrole_counts.index, y=jobrole_counts.values)
plt.xlabel('Job Role')
plt.ylabel('Number of Employees')
plt.title('Distribution Across Job Roles')
plt.xticks(rotation=90)

plt.tight_layout() 
plt.show()
```

```python
plt.figure(figsize=(12,5))

# Histogram for Age
plt.subplot(1, 2, 1)
sb.histplot(df['Age'], bins=10, kde=True)
plt.xlabel('Age')
plt.ylabel('Employees')
plt.title('Age Distribution')

# Histogram for Monthly Income
plt.subplot(1, 2, 2)
sb.histplot(df['MonthlyIncome'], bins=10, kde=True)
plt.xlabel('Monthly Income')
plt.ylabel('Employees')
plt.title('Monthly Income Distribution')

plt.show()
```

# CORRELATION ANALYSIS

```python
# Compute the correlation matrix
corr_matrix = df.corr(numeric_only=True)
# Plot the heatmap
plt.figure(figsize=(8,6))
sb.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#Compare Satisfaction, Salary & Performance Using Box Plots

plt.figure(figsize=(15, 5))  

# Job Satisfaction
plt.subplot(1, 3, 1)
sb.boxplot(x=df['Attrition'], y=df['JobSatisfaction'])
plt.title("Job Satisfaction vs Attrition")

# Monthly Income
plt.subplot(1, 3, 2)
sb.boxplot(x=df['Attrition'], y=df['MonthlyIncome'])
plt.title("Salary vs Attrition")

# Performance Rating
plt.subplot(1, 3, 3)
sb.boxplot(x=df['Attrition'], y=df['PerformanceRating'])
plt.title("Performance Rating vs Attrition")

plt.tight_layout()
plt.show()
```

```python
#Identify Attrition Trends Across Demographics

# Set the figure size
plt.figure(figsize=(15, 8))

# Attrition by Gender
plt.subplot(2, 2, 1)
sb.countplot(data=df, x="Gender", hue="Attrition")
plt.title('Attrition by Gender')

# Attrition by Age
plt.subplot(2, 2, 2)
sb.countplot(data=df, x="Age", hue="Attrition")
plt.xticks(rotation=90)
plt.title('Attrition by Age')

# Attrition by Department
plt.subplot(2, 2, 3)
sb.countplot(data=df, x="Department", hue="Attrition")
plt.title('Attrition by Department')

# Attrition by Education Field
plt.subplot(2, 2, 4)
sb.countplot(data=df, x="EducationField", hue="Attrition")
plt.title('Attrition by Education Field')

# Display the plots
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))

# Attrition by Department
plt.subplot(1,2,1)
sb.countplot(x='Department', hue='Attrition', data=df)
plt.title("Attrition by Department")
plt.xticks(rotation=30)

# Attrition by Job Role
plt.subplot(1,2,2)
sb.countplot(x='JobRole', hue='Attrition', data=df)
plt.title("Attrition by Job Role")
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
